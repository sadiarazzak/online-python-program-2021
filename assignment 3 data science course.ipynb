```python
data["cast"].replace(np.nan, "None",inplace=True)
```

```python
data["cast"].isnull().sum()
```

```python
def num_cast(val):
    if val== "None":
        return 0
    else:
        return len(str(val).split(", "))
```

```python
data["No of cast"]=data["cast"].apply(num_cast)
data.head()
```
